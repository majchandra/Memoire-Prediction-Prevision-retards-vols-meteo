## Data Extraction

```python
# Extract train and validation loss
train_loss = []
val_loss = []

# Iterating through each line of data containing the loss values for each epoch
for line in data_loss.splitlines():
    
    # Separate the line at each comma to separate the epoch, the train loss and the validation loss
    line_splitting = line.split(',')
    
    # Extract the train loss value
    train_loss.append(float(line_splitting[1].split(': ')[1]))

    # Extract the validation loss value
    val_loss.append(float(line_splitting[2].split(': ')[1]))
```

```python
# Extract the confusion matrix
confusion_matrix = ','.join(data_confusion_matrix.split())
confusion_matrix = confusion_matrix.replace('[,', '[')
confusion_matrix = np.array(eval(confusion_matrix))
```

## Display graphs

```python
# Plotting training and validation loss
epochs = range(1, len(train_loss) + 1)

plt.figure(figsize=(10, 6))

plt.plot(epochs, train_loss, label='Train Loss')
plt.plot(epochs, val_loss, label='Validation Loss')

# Axis configuration
plt.xlabel('Epochs', fontweight='bold', fontsize=14)
plt.ylabel('Loss', fontweight='bold', fontsize=14)
plt.xticks(fontsize=13)
plt.yticks(fontsize=13)

# Define grid 
plt.grid(True)

# Define plot legend
plt.legend()

plt.show()
```

```python
# Display the confusion matrix
plt.figure(figsize=(10, 6))
sns.heatmap(confusion_matrix, annot=True, fmt='d', cmap='Blues')

# Axis configuration
plt.xlabel('Predicted',fontweight='bold', fontsize=14)
plt.ylabel('Actual', fontweight='bold', fontsize=14)
plt.xticks(fontsize=13)
plt.yticks(fontsize=13)

plt.show()
```
